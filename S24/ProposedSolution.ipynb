```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
csv_file = 'CampioneA.csv'
df = pd.read_csv(csv_file, sep=';')
```

```python
date_locked = df['Date'][0]
isnan = df['Date'].isnull()
```

```python
# to fill NaN (is unefficient)
for i in range(len(df)):
    if isnan[i] == True:
        df['Date'][i] = date_locked
    else:
        date_locked = df['Date'][i]
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Day'] = df['Date'].apply(lambda x: x.day)
df['Month'] = df['Date'].apply(lambda x: x.month)
df['Year'] = df['Date'].apply(lambda x: x.year)
df['Year'] = df['Date'].apply(lambda x: x.year)
df['Day_Name'] = df['Date'].apply(lambda x: x.day_name())
df
```

```python
df_groupedbyday = df.groupby(df['Date']).sum()
df['Sum orders 18h'] = df['Orders placed'].rolling(18).sum()
df_groupedbyday
# 16/05 is the day with most completed orders
```

```python
print(df['Sum orders 18h'].max())
df[df['Sum orders 18h'] == df['Sum orders 18h'].max()]
# 2nd answer 378.0 orders 18h ending the 17/05 at 3 (saturday-Sunday)
```

```python
df_groupedbyhour = df.groupby(df['Time']).sum()
df_groupedbyhour
# explore datas aggregated by hour
```

```python
list_of_hours = [5,11,18,14]
df_groupedbyhour.loc[list_of_hours]
# 3nd answer is 4-5 (the hour with less orders in the whole period from the options)
```
